## For Loops, Conditional, Lists, Ranges




For-loops can be used to iterate numerical values like in other programming languages with the range function:

````python
for i in range(0, 10):
    print(i)
````

But can also be used to visit every item in a list.

````python
for color in ["red", "green", "blue"]:
    print(color)
````

Remember that the contents of the for-loop have to be indented at the same level to differentiate them from code outside the for-loop:

````python
for i in range(3):
    print("repeated")
    print("also repeated")
print("not repeated")
````

## Dictionary

A dictionary is a collection which is unordered, changeable and indexed. In Python dictionaries are written with curly brackets, and they have keys and values. Dictionaries can be created in a variety of ways:

````python
my_dict = {}   # Empty dict
my_dict = {'foo': 'bar', 'baz': 'bak'}
print(my_dict)
# This one is handy if you have a list of pairs to turn into a dictionary:
your_dict = dict([['foo', 'bar'], ['baz', 'bak']])
print(your_dict)
````

`'foo'` and `'baz'` are called keys, `'bar'` and `'bak'` are called values. You can access values in the dictionary with its key:

````python
my_dict['foo']
````

And you can add new values (or overwrite old ones) by key as well:

````python
my_dict['hello'] = 'world'
my_dict['hello'] = 'goodbye'
print(my_dict)
````

````python
my_dict['last'] = 80
````

You can iterate over a dictionary using a for-loop:

````python
for key in my_dict:
    print("The key", key, "maps to the value", my_dict[key])
    #print("The key " +  key + " maps to the value " + my_dict[key])
````

````python
list(my_dict.keys())
````

````python
list(my_dict.values())
````

````python
my_dict.items()
````

````python
del my_dict['hello']
my_dict
````

##Functions and Map

You can define your own functions using the `def` keyword and `return` to specify the value that is returned by the function.

````python
def double_plus_y(x, y=4):
    return 2 * x + y

double_plus_y(6, 9)
````

But functions don't have to take parameters (`x` and `y` in the example above) or return anything:

````python
def say_hi():
    print("Just saying 'hello'.")

say_hi()
````

Another function has default parameters.

````python
def my_function(x, y, z=1.5):
    if z > 1:
        return z * (x + y)
    else:
        return z / (x + y)
````

````python
print(my_function(5, 6, z=0.7))
print(my_function(3.14, 7, 3.5))
my_function(10, 20)
````

Functions can return multiple values.

````python
def f():
    a = 5
    b = 6
    c = 7
    return a, b, c

a, b, c = f()
````

````python
return_value = f()
return_value
````

The map function allows us to call a function on each item in a list:

```
# This is formatted as code
def double_plus_y(x, y=4):
    return 2 * x + y
```



````python
my_list
````

````python
for value in map(double_plus_y, my_list):
    print(value)
````

````python
new_lst = list(map(double_plus_y, my_list))
new_lst
````

For simple, one-time-use function, we don't have to define a function, we can use lambda to define the operation in-line:

````python
for value in map(lambda x: 2 * x, my_list):  # Don't need a separate function
    print(value)
````

Note that lambda functions don't use the `return` keyword, you just specify the names of the parameters of the function (`x` in the example above), a colon, and the operation to perform on the parameter(s).



## More on List Sorting, Slicing, and List Comprehension

Sorting a list with different order property.

````python
a = [7, 2, 5, 1, 3]
a.sort() # in-place sort
a
````

````python
b = ['six', 'small', 'He', 'foxes', 'saw']
b.sort(key=len)
b
````

Slicing a list in various ways by specifying indexes.

````python
seq = [7, 2, 3, 7, 5, 6, 0, 1]
seq[1:5]
````

````python
seq[0:4] = [9, 9] # insert new elements to replace current elements in A range
print(seq)
print(seq[:5])
print(seq[2:])
print(seq[-4:])# counting from end
print(seq[-6:-2])
seq[::2] # slicing by the step of 2
seq[::-1] # slicing by the step of 1, reverse
````

Use [list comprehension](https://www.pythonforbeginners.com/basics/list-comprehensions-in-python) to perform an operation on every item in the list. It looks a little bit like a for-loop inside of a list:

````python
[x*2 for x in my_list]
````

You can also use it to filter out values from a list. For example to extract every odd values from the list:

````python
[x for x in my_list if x % 2 == 1]
````

You can even combine filtering and other operations:

````python
[x**2 for x in my_list if x<10]   # Square every value less than 10
````

Let's see a few more complex examples for list comprehensions. To do so, we're going to be using a list of city and state names. Fun fact: these are all [real cities](https://en.wikipedia.org/wiki/List_of_the_most_common_U.S._place_names) in the US but with a more famous namesake in a different state.



````python
cities = ['washington,ct', 'springfield,or', 'riverside,tx', 'franklin,vt', 'lebanon,co', 'dayton,tx', 'las vegas,nm', 'madison,ca', 'georgetown,ct', 'los angeles,tx']
````

**Use list comprehension to produce a list of only the cities whose name (including the state name) are less than 12 characters long.**

````python
[city  for city in cities if len(city) < 12]
````

**Create a list of abbreviations that are just the first 3 letters of each city name:**

````python
[city[:3]  for city in cities]
````

````python
cities
````

**Use list comprehension, to create a dictionary that maps city names to the states that they are located in.**

````python
city_lst = [city.split(',')[0] for city in cities]
print(city_lst)
st_lst = [city.split(',')[1] for city in cities]
print(st_lst)
city_dict = dict(zip(city_lst, st_lst))
city_dict
````
