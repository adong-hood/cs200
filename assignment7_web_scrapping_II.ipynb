```python
# This web archive link is slow and not always reliable.
#address = 'https://web.archive.org/web/20260203121119/https://www.cia.gov/the-world-factbook/field/infant-mortality-rate/'
#- Insteadn, use this link for 2024 rates.
address_2024 = 'http://pluto.hood.edu/~dong/datasets/factbook/fields/infantmortality-2024.html'
res = requests.get(address_2024)
print(res.status_code)
page_2024 = res.text
soup_2024 = BeautifulSoup(page_2024)
all_countries = soup_2024.select('.pb30')
infant_mortality_df_2024 = {}
for country in all_countries:
  country_name = country.find('h3').get_text()
  total_rate = country.find('p').text.strip().split(":")[1].strip().split('deaths/')[0].strip()
  # print(country_name)
  # print(total_rate)
  infant_mortality_df_2024[country_name] = float(total_rate)

infant_mortality_df_2024
infant_2024_df = pd.DataFrame({'infant-mortality-rate-2024':infant_mortality_df_2024})
infant_2024_df.head()
```
