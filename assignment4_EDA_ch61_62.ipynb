## Case Study 1: World Factbook Dataset

Here we'll just try to use the data set to answer some simple questions about the countries of the world.

```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt

#-%matplotlib inline
plt.style.use('classic')
#-import numpy as np
```

```python
wd_df = pd.read_csv('http://pluto.hood.edu/~dong/datasets/world_countries.csv')
#- basic check with shape, column, head, describe, info
```

1. As a warmup, find the top 10 countries with the most population.

```python
wd_df.sort_values(by = 'Population', ascending=False, inplace=False)[['Country']].head(10)
```

2. Find out the number of countries in each region. <br />[Use group by with dataframe.](https://jakevdp.github.io/PythonDataScienceHandbook/03.08-aggregation-and-grouping.html)


```python
wd_df.Region.value_counts()
```

```python
wd_df.groupby('Region').agg({"Country":'count'})
#- wd_df.groupby('Region')[["Country"]].count()
```

<font color = "green">Visualizing Distribution with Histograms</font>

3. What does the distribution of GDP look like?

```python
alt.Chart(wd_df).mark_bar().encode(x=alt.X('Birthrate', bin=True), y='count()')
alt.Chart(wd_df).mark_bar().encode(x=alt.X('Birthrate', bin=alt.Bin(maxbins=20)), y='count()') #- bin number ignored.
```

```python
plt.hist(wd_df['Birthrate'], bins = 20)
plt.show()
```


Why is that graph so skewed? Do you think it will change if you scale GDP by population?

4. Let's scale GDP by population and add it as a new coulum to the dataframe.

```python
wd_df['GDP_scaled_by_pop'] = wd_df['GDP']/wd_df['Population']
```

The distribution of GDP_scaled_by_pop is also vey skewed, so population does not explain everything!

<font color = "green"> Scatter Plots for discovering relationships </font>

5. What is the relationship between literacy rate and infant mortality? Why do you think that is?


```python
alt.Chart(wd_df).mark_point().encode(x='Population', y='Area', tooltip='Country', color = 'Region')
```

6. How does that relationship between literacy rate and infant mortality change when you look at different regions?

```python
plt.scatter(wd_df['Population'], wd_df['Area'])
plt.show()
```

<font color = "green">Power Tools: Scatter Matrix</font>

It would be pretty tedius to look at all the different pairs of things we might want to look at for correlation one at a time, but we can use a scatter matrix to make life easier.

```python
alt.Chart(wd_df).mark_circle().encode(                      #- Scatterplot circles from wd_df
    alt.X(alt.repeat("column"), type='quantitative'),       # X = repeated numeric column
    alt.Y(alt.repeat("row"), type='quantitative'),          # Y = repeated numeric row
    color='Region:N',                                       # Color by region (categorical)
    size='Population:Q',                                    # Size by population (numeric)
    tooltip='Country:N'                                     # Show country on hover
).properties(
    width=150,                                              # Plot width
    height=150                                              # Plot height
).repeat(
    row=['Birthrate', 'Deathrate', 'Infant mortality', 'GDP'],      # Variables on rows
    column=['Birthrate', 'Deathrate', 'Infant mortality', 'GDP']    # Variables on columns
).interactive()  # Enable zoom/pan
```
