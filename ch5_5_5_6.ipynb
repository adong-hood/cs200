#### 3. Find the population for each country.

Frequently data come from different sources and have to be merged into a single data frame. More examples can be found from this site: <a href = "https://jakevdp.github.io/PythonDataScienceHandbook/03.07-merge-and-join.html">Combining Datasets: Merge and Join</a>

#### 4. Add notes to some movies.
let's say that I have some notes about some of these movies that I want to merge:

```python
my_notes_dict = {
    "Cutthroat Island": "Has one of my favorite stunts",
    "The Neverending Story III: Escape from Fantasia": "Too many sequels here",
    "Bio-Dome": "First Pauly Shore movie I ever saw",
    "The Empire Strikes Back": "My favorite in the SW series",
    "Mighty Aphrodite": "Features Helena Bonham Carter",
}
my_notes = pd.DataFrame(pd.Series(my_notes_dict), columns=['my_notes'])
my_notes['title'] = my_notes.index
my_notes
```

```python
merged_df = pd.merge(movie_df, my_notes, on='title', how='inner')[['title', 'my_notes', 'vote_count']]
merged_df.head()
```

```python
merged_df['double_vote_count'] = merged_df['vote_count'] * 2
merged_df.head()
```
