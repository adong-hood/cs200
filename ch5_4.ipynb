### 4.4 Find the number of country names starting with B or C.
Slicing based on string comparison.

```python
happiness_df.index
```

```python
lookup_by_country = happiness_df.set_index('Country').sort_index()
lookup_by_country.head(2)
```

```python
lookup_by_country.index
```

```python
begin = lookup_by_country[lookup_by_country.index.str.startswith('B')].index[0]
```

```python
end = lookup_by_country[lookup_by_country.index.str.startswith('C')].index[-1]
```

```python
lookup_by_country.loc[begin:end].shape
```

We can use the `startswith` predicate (in the `str` attribute) to find all countries that start with a particular string or letter.
